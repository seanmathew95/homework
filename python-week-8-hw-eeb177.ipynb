```python
from collections import defaultdict
species_ranges = defaultdict(list)
canids = open("../class-assignments/canids/formatted_canids.csv")
canids.readline()
canid_dict= {}
for line in canids:
    species = line.split('","')[5]
    min_ma = float(line.split('","')[11])
    max_ma = float(line.split('","')[10])
    avg_ma = (max_ma + min_ma )/ 2
    
    species_ranges[species].append(avg_ma)
canids.close()


```

```python
#get the oldest and youngest for each species
output = open("canid_ranges.csv", "w")
alphabetical = sorted(list(species_ranges.keys()))
for key in alphabetical:
    species = key
    ages = species_ranges[key]
    minage = min(ages)
    maxage = max(ages)
    genus = key.split(" ")[0]
    #make sure no empty species are in data
    if genus != species:
        print(genus, species, minage, maxage)
        output.write(genus + "," + species + "," + str(minage) + "," + str(maxage) + "\n")
output.close()
```

### Write another loop that reports the number of species in each family in the ebird data set with this output "There areXX total species in Family YY"
import re
from collections import Counter
ebird_file=open("/home/eeb177-student/Desktop/eeb-177/homework/data-hw5/formatted-ebird.csv")

for line in ebird_file:
    csv=re.split(",", line)
    family=csv[7]
    ebird_list.append(family)
counted = Counter(ebird_list)

for key in counted:
    print("There are " + str(counted[key]) + " total species in Family " + key)

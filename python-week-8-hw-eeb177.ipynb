```python
from collections import defaultdict
species_ranges = defaultdict(list)
canids = open("../class-assignments/canids/formatted_canids.csv")
canids.readline()
canid_dict= {}
for line in canids:
    species = line.split('","')[5]
    min_ma = float(line.split('","')[11])
    max_ma = float(line.split('","')[10])
    avg_ma = (max_ma + min_ma )/ 2
    
    species_ranges[species].append(avg_ma)
canids.close()


#get the oldest and youngest for each species
output = open("data_hw8/canid_ranges.csv", "w")
alphabetical = sorted(list(species_ranges.keys()))
for key in alphabetical:
    species = key
    ages = species_ranges[key]
    minage = min(ages)
    maxage = max(ages)
    genus = key.split(" ")[0]
    #make sure no empty species are in data
    if genus != species:
        output.write(genus + "," + species + "," + str(minage) + "," + str(maxage) + "\n")
output.close()
```

### Functions

```python
def extract_fossil_ages(filename):
    from collections import defaultdict
    species_ranges = defaultdict(list)
    genera = open(filename)
    genera.readline()
    genera_dict= {}
    for line in genera:
        species = line.split('","')[5]
        min_ma = float(line.split('","')[11])
        max_ma = float(line.split('","')[10])
        avg_ma = (max_ma + min_ma )/ 2
        genus = species.split(" ")[0]
        if species != genus:
            species_ranges[species].append(avg_ma)
    genera.close()
    return species_ranges

extract_fossil_ages("../class-assignments/canids/formatted_canids.csv")
```

```python
def dict_to_file(species_ranges_dict, new_filename):
    output_file = open(new_filename, "w")
    alphabetical = sorted(list(species_ranges_dict.keys()))
    for key in alphabetical:
        species = key
        
        ages = species_ranges_dict[key]
        minage = min(ages)
        maxage = max(ages)
        genus = key.split(" ")[0]
        output_file.write(genus + "," + species + "," + str(minage) + "," + str(maxage) + "\n")
    
    output_file.close()
    return output_file

canid_ranges = extract_fossil_ages("../class-assignments/canids/formatted_canids.csv")
dict_to_file(canid_ranges, "data_hw8/canid_ranges.csv")
```

```python
def get_longest_species_range(input_dict):
    species_ages = {}
    for key in input_dict.keys():
        ages = input_dict[key]
        min_age = min(ages)
        max_age = max(ages)
        age_range = max_age - min_age
        species_ages[key] = age_range
    highest_age = max(list(species_ages.values()))
    for key in species_ages.keys():
        if species_ages[key] == highest_age:
            longest_range_species = key
    return longest_range_species

get_longest_species_range(canid_ranges)
```

```python
def extract_genus_ages(filename):
    from collections import defaultdict
    genus_ranges = defaultdict(list)
    genera = open(filename)
    genera.readline()
    genera_dict= {}
    for line in genera:
        species = line.split('","')[5]
        genus = species.split(" ")[0]
        if genus !=species:
            min_ma = float(line.split('","')[11])
            max_ma = float(line.split('","')[10])
            avg_ma = (max_ma + min_ma )/ 2
            genus_ranges[genus].append(avg_ma)
    genera.close()
    return genus_ranges



def genus_dict_to_file(genus_ranges_dict, new_filename):
    output_file = open(new_filename, "w")
    alphabetical = sorted(list(genus_ranges_dict.keys()))
    for key in alphabetical:
        genus = key
        ages = genus_ranges_dict[key]
        minage = min(ages)
        maxage = max(ages)
        output_file.write(genus + "," + str(minage) + "," + str(maxage) + "\n")
    
    output_file.close()
    return output_file

def get_longest_genus_range(genus_range_dict):
    genus_ages = {}
    for key in genus_range_dict.keys():
        max_age = max(list(genus_range_dict[key]))
        min_age = min(list(genus_range_dict[key]))
        age_range = max_age - min_age
        genus_ages[key] = age_range
    max_range = max(list(genus_ages.values()))
    for key in genus_ages.keys():
        if genus_ages[key] == max_range:
            max_genus = key
    return max_genus


feliform_genus = extract_genus_ages("data_hw8/formatted_feliform.csv")

genus_dict_to_file(feliform_genus, "data_hw8/feliform_genus.csv")

get_longest_genus_range(feliform_genus)
```
